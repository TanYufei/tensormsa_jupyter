```python
import tensorflow as tf
```

```python
x = tf.constant([-2.25, -3.25])
y = tf.assert_negative(x)
```

```python
xp = tf.constant([-2.25, 3.25])
yp = tf.assert_positive(xp)
```

```python
with tf.Session() as session:
    print("y ",type(y))
    print(y.run())
    print("yp",type(yp))
    print(yp.run())
```
